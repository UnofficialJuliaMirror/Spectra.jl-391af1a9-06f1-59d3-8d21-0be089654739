```julia
using Spectra
using PyPlot
```

```julia
# the x axis
x = collect(0:0.1:150)

# a scale factor that you can change
scale = 0.1

# a fake signal: perfect y
y_tot, y_peaks = gaussiennes([10;20.],[40.;60],[5.;15],x)
bkg, ~ = gaussiennes([10.],[50.],[50.],x)
y_perfect = scale.*(y_tot+bkg)

# we add noise: observed y
y = scale.*(y_tot+bkg + randn(size(y_tot,1)))

roi = [0 20.; 100. 150]

interest_index = find(roi[1,1] .<= x[:,1] .<= roi[1,2])
if size(roi)[1] > 1
    for i = 2:size(roi)[1]
        interest_index = vcat(interest_index,  find(roi[i,1] .<= x[:,1] .<= roi[i,2]))
    end
end

x_bas = x[interest_index]
y_bas = y[interest_index]

figure()
plot(x,y,"k-",label="Noisy observed signal")
plot(x_bas,y_bas,"r.",label="ROI")
xlabel("Position, a. u.",fontname="Sans",fontsize = 18)
ylabel("Amplitude, a. u.", fontname = "Sans", fontsize = 18)
legend()
```
